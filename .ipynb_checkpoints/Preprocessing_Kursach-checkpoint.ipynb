```python
import numpy as np
import random
def transition(transition_probabilities, rating):
    intervals = np.cumsum(transition_probabilities[rating, :])
    eps = random.uniform(0, 1)
    if eps < intervals[0]:
        return 0
    if eps >= intervals[0] and eps < intervals[1]:
        return 1
    if eps >= intervals[1] and eps < intervals[2]:
        return 2
    if eps >= intervals[2] and eps < intervals[3]:
        return 3
    if eps >= intervals[3]:
        return 4
```

## Непрерывный случай

```python
import numpy as np
def EAD(amount, interest, number_of_payments, time):
    main_payment = amount / number_of_payments
    i = time // 0.25
    terms = time % 0.25
    amount -= i * main_payment
    interest_payment = amount * terms * (interest / 100)
    EAD = amount + interest_payment
    return EAD

def transition_continuous(transition_probabilities, amount, rating, time = 1):
    losses = 0
    change_time = np.random.exponential(-1 / transition_probabilities[rating, rating])
    if change_time < time:
        proba = transition_probabilities / np.diagonal(- transition_probabilities).reshape((4, 1))
        np.fill_diagonal(proba, 0)
        proba = proba / proba.sum(axis = 1).reshape((4, 1))
        choice = np.random.choice(range(5), p = proba[rating, :])
        if choice == 4:
            losses += EAD(amount, 10, 4, change_time)
    return losses
```

```python
import numpy as np
import matplotlib.pyplot as plt
EADs = []
time = []
for t in np.arange(0, 1.001, 0.001):
    EADs.append(EAD(1000, 10, 4, t))
    time.append(t)

plt.plot(time, EADs)
plt.axvline(0.25, linestyle = "dotted", color = "red")
plt.text(0.2, 400, "Выплата", rotation = "vertical")
plt.xlabel("Время")
plt.ylabel("EAD")
```
